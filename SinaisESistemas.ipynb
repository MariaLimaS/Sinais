```python
import numpy as np
import matplotlib.pyplot as plt

# Loop para criar os gráficos de 1 a 5
for i in range(1, 6):
    N = i
    Fs = 8192
    t = np.linspace(0, 1 - 1 / 8192, Fs)
    f = 2
    amp = 1
    plt.subplot(3, 2, N)  # Cria um subplot na posição N
    h, = plt.plot(np.nan, np.nan)  # Inicializa um objeto de linha vazio
    tt = 'Ordem {}'.format(N)
    plt.title(tt)
    plt.axis([0, 1, -0.2, 1.2])  # Define os limites dos eixos
    res = amp / 2
    
    # Loop para calcular a forma de onda do seno
    for k in range(1, N + 1):
        res = res - (amp / (k * np.pi)) * np.sin(2 * np.pi * k * f * t)
        h.set_xdata(t)  # Atualiza os dados do eixo x do objeto de linha
        h.set_ydata(res)  # Atualiza os dados do eixo y do objeto de linha
        plt.draw()  # Redesenha o gráfico
        #plt.pause(0.1)  # Pausa para exibir o gráfico gradualmente

    plt.show(block=False)  # Exibe o gráfico sem bloquear a execução
    plt.pause(0.1)  # Pausa para exibir o gráfico por um curto período
    plt.clf()  # Limpa a figura atual para criar uma nova na próxima iteração

# Configurações para o último gráfico com N=500
N = 500
Fs = 8192
t = np.linspace(0, 1 - 1 / 8192, Fs)
f = 2
amp = 1
plt.subplot(3, 2, 6)
h, = plt.plot(np.nan, np.nan)
tt = 'Ordem {}'.format(N)
plt.title(tt)
plt.axis([0, 1, -0.2, 1.2])
res = amp / 2

# Loop para calcular a forma de onda do último gráfico
for k in range(1, N + 1):
    res = res - (amp / (k * np.pi)) * np.sin(2 * np.pi * k * f * t)
    h.set_xdata(t)
    h.set_ydata(res)
    plt.draw()
    #plt.pause(0.1)

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

f = 10
overSampRate = 30
fs = overSampRate * f
nCyl = 5
t = np.arange(0, nCyl * 1 / f, 1 / fs)
x = np.sin(2 * np.pi * f * t)

plt.plot(t, x)  # Plota o gráfico de x em função de t
plt.title('Seno f={}Hz'.format(f))  # Define o título do gráfico
plt.xlabel('Tempo(s)')  # Define o rótulo do eixo x
plt.ylabel('Amplitude')  # Define o rótulo do eixo y
plt.xlim(0, 0.5)  # Define os limites do eixo x
plt.ylim(-1, 1)  # Define os limites do eixo y
plt.show()  # Exibe o gráfico
```
